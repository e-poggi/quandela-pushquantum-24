```python
import perceval as pcvl
import perceval.components.unitary_components as pcvl_comp
import numpy as np
from collections import Counter
pcvl.__version__
```

```python
zero_state=pcvl.StateVector([0, 1, 0, 0, 1, 0])
psi_state=pcvl.StateVector([0, 0, 1, 1, 0, 0])
```

```python

circuit = pcvl.Circuit(6)  # Create a 6 mode circuit

H = pcvl_comp.BS.H
Phi = lambda angle: pcvl_comp.PS(phi=pcvl.P(angle))

circuit.add(2, H())
circuit.add(2, Phi('phi1'))
circuit.add(2, H())
circuit.add(3, Phi('phi2'))

circuit.add(1, H()).add(3, H())
circuit.add(1, Phi('theta1')).add(3, Phi('theta2'))
circuit.add(1, H()).add(3, H())
circuit.add(2, Phi('theta3')).add(4, Phi('theta4'))
circuit.add(2, H())
circuit.add(2, Phi('theta5'))
circuit.add(2, H())
circuit.add(1, Phi('theta6')).add(3, Phi('theta7'))
circuit.add(1, H()).add(3, H())
circuit.add(1, Phi('theta8')).add(3, Phi('theta9'))
circuit.add(1, H()).add(3, H())
circuit.add(2, Phi('theta10')).add(4, Phi('Phi11'))
circuit.add(2, H())
circuit.add(2, Phi('theta12'))
circuit.add(2, H())


circuit.add(1, Phi('phi3')).add(3, Phi('Phi4'))
circuit.add(1, H()).add(3, H())
circuit.add(1, Phi('phi5')).add(3, Phi('Phi6'))
circuit.add(1, H()).add(3, H())


pcvl.pdisplay(circuit)
```
